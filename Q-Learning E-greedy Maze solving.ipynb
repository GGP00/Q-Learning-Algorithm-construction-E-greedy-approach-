This grid represents the **environment**. Every environment in a Reinforcement Learning problem must be always structured as sequential sets of alternatives
from which to infer an optimal combination. Each of these alternatives are states. It also
has information related to every single state, which must at least contain an associated
reward. In other words, the environment has to fit any model which follows a **Markov
Decision Process (MDP)**. Additionally, the agent must be given a certain initial
state from which to start the transition process. It ends when the final state is reached or
interrupted by an external event.

<h3> The memory of our agent (q_table) will work as a dictionary key=State: value=Action's suitability.</h3>

Remember that "to make a class hashable, it has to implement both the __hash__(self) method and the aforementioned __eq__(self, other) method. Sets and dictionaries use hashes for their lookup tables to quickly find their keys. If an attribute can change in the lifetime of an object, you can’t use it for hashing. Why the equality check? As we’ve established before, a hash is an integer. And even though we have 64 bits to splurge on modern architectures, there’s still the possibility that two objects have the same hash." quoted by https://hynek.me/articles/hashes-and-equality/

```python
class State:
    
    def __init__(self, grid, agent_pos):
        self.grid = grid
        self.agent_pos = agent_pos
    
    def __eq__(self, other):
        return isinstance(other, State) and self.grid == other.grid and self.agent_pos == other.agent_pos
    
    def __hash__(self):
        return hash(str(self.grid) + str(self.agent_pos))
    
    def __str__(self): 
        return f"State(grid={self.grid}, agent_pos={self.agent_pos})" 

q_table = dict() # Our agent's brain!
```

<h3> In each iteration (step) the agent moves to a new position and 'experience' a positive or negative feedback relating to the item he finds (reward). </h3>

The reward_value dictionary contains the string simbol of the desire item as key, and a list of [ reward for that item, True/False if the agent ends the game when steping into that item ] values. Further, in the observe_value function this will be iterated to compare and extract the current slamed cell item.

The observe_reward_value function has been designed for admiting wether string with their respective associated rewards, or directly the rewards stored as float type.


```python
from copy import deepcopy

reward_value = {'*': [-1, False], 'f': [-10, False], 'X': [100, True] }

def observe_reward_value(state, action):    
    x,y = action
    grid_item = state.grid[x][y]          
    new_grid = deepcopy(state.grid) # in case you want to modify the current grid
    values = reward_value.get(grid_item, f"Unknown grid item {grid_item} appeared in cell {action}")
    if isinstance(grid_item, float) is True:
            reward = grid_item 
            is_done = False
    elif isinstance(values, str) is True:
        raise ValueError(f"{values}")
    else:
        reward, is_done = values
    
    return State(grid=new_grid, agent_pos=(x,y)), reward, is_done
```

<h3> << WHEN ESCALATING GRID'S SIZES !! >> </h3>It is very important to adjust the N_EPISODES according to the amount of cells which make up the designed grid in order to avoid inconsistent results, such as unexplored directions (0. quality value) or not optimous solutions in terms of distance. Aproximately N_EPISODES should be from 4 to 7 times the number of cells, depending on the number of possible actions enabled to the agent. 

```python
import numpy as np
import random

random.seed(42) # for reproducibility

n_episodes = 5 * len(grid) * len(grid[0])
MAX_EPISODE_STEPS = 100 
min_alpha = 0.02

alphas = np.linspace(1.0, min_alpha, n_episodes) # Probability decays while steps get higher, not episodes! (the path follow to reach a determinate reward becames less and less likely in each step)
gamma = 1.0

eps = 1.0 # Exploration rate (changes inversely proportional to the number of episodes, not steps!)
max_epsilon = 1.0  # Exploration probability at start (maximum)
min_epsilon = 0.01 # Minimum exploration probability
decay_rate = 0.01 # Exponential decay rate for exploration probability
```

The function below depends on the type of the problem dealt with. The possible actions represent the enabled behaviour or interaction of our agent with the created environment. For instance, in the current case the actions represent the steps that the agent can do in every single direction. 

```python
import numpy as np

possible_actions = []
next_possible_actions = []

def extract_possible_actions(state):
    
    x,y = state.agent_pos
    around_area = [(x, y-1), (x, y+1), (x-1, y), (x+1, y), (x+1,y+1), (x-1,y-1), (x+1,y-1), (x-1,y+1)]
    possible_actions.clear()

    for _x, _y in around_area:
            if _x > (len(state.grid) - 1) or _y > (len(state.grid[0]) - 1): # not crossing boundaries of the grid
                continue
            if grid[_x][_y] == 'o': # not in obstacles
                continue
            if _x == x and _y == y: # going backwards is useless
                continue
            possible_actions.append((_x, _y))
    #print(f"extract_possible_actions from state.pos: {state.agent_pos} => {possible_actions}")
    return possible_actions
        
```

The memorization process is based on the **exploration vs exploitation** dilemma:
Let's imagine a man lost in the jungle. In order to survive the man decides to explore his surroundings in search of feeding resources. As the man walks through the jungle and generates experience, he learns how to
avoid dangers while successfully finding food. As a rational being, he starts to exploit that
knowledge, that is, he utilizes the optimal policy in order to maximize the environment's
reward. To sum up:
* Exploration: random choice of any of all the possible actions a in state s.
* Exploitation: choice of maximum action's quality value for all possible actions a in state s

```python
def choose_action(state):
    
    if random.uniform(0, 1) < eps: 
        return random.choice(list(range(len(q_table[state]))))
    else: 
        return np.argmax(q_table[state])
```

The construction of the **optimal policy** (or escaping path) is immediate once all possible
combinations of **Q(s; a)** (quality values per single state and each of its respective actions) have converged to an accurate representative value **Q**. This can be made by selecting the maximum Q(s; a) in each state, starting from the one given.

```python
def inference_path ():
    state = start_state
    path = []
    for _ in range(MAX_EPISODE_STEPS):
        action = np.argmax(q_table[state])
        new_state, reward, done= observe_reward_value(state, extract_possible_actions(state)[action])
        state = new_state
        path.append(new_state.agent_pos)
  
        if new_state.grid[new_state.agent_pos[0]][new_state.agent_pos[1]] == 'X':
            print(f"EXIT 'X' REACHED THROUGTH: {path}")   
            break
    return path
```

<h3> Memorization process: n_episodes 'games' with a maximum of 100 actions done by the agent. </h3> 

This is the main function where all the logistics described above come together.
_Don't miss the printed information below this cell code while the learning process is being runned._

This algorithm is called Bellman's equation and represents the agent's brain. In other words, the agent is an entity with cognitive abilities (memory and deduction). Identifying the agent in a RL problem is simple, it is the element that owns conduct (defined by a certain set of actions) and on which the optimal execution of the problem to be solved
depends. In this case it represents a man escaping from fire.

```python
def learn (start_state):

    for e in range(n_episodes): 
        state = start_state 
        total_reward = 0
        alpha = alphas[e]

        for _ in range(MAX_EPISODE_STEPS):
            possible_actions = extract_possible_actions(state)
            if state not in q_table:
                q_table[state] = np.zeros(len(possible_actions))
            action = choose_action(state) 
            next_state, reward, done = observe_reward_value(state, possible_actions[action])
            total_reward += reward 
            if next_state not in q_table:
                q_table[next_state] = np.zeros(len(extract_possible_actions(next_state)))
            q_table[state][action] = q_table[state][action] + \
                     alpha * (reward + gamma *  np.max(q_table[next_state]) - q_table[state][action])
            state = next_state
            if done:
                break

        eps = min_epsilon + (max_epsilon - min_epsilon)*np.exp(-decay_rate*e) 

        print(f"Episode {e}/{n_episodes-1} with total reward {total_reward} ", end="\r", flush=True)

learn(start_state=start_state)
```
